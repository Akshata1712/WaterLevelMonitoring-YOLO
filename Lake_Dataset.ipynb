```python
import cv2
import os
```

```python
model_path = "/content/drive/MyDrive/yolo_training/exp1/weights/best.pt"  # trained model
source_folder = "/content/drive/MyDrive/Camera lake_frames 2 timestamp"   # folder with input images
save_folder = "/content/drive/MyDrive/Cropped_Lake_Images"
```

```python
os.makedirs(save_folder, exist_ok=True)

model = YOLO(model_path)
```

```python
for img_name in os.listdir(source_folder):
    img_path = os.path.join(source_folder, img_name)

    # Skip non-images
    if not img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    # Run inference
    results = model(img_path)

    # Read original image
    img = cv2.imread(img_path)

    # Process detections
    for r in results:
        boxes = r.boxes.xyxy  # bounding boxes: [x1, y1, x2, y2]
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box)
            crop = img[y1:y2, x1:x2]

            # Keep the SAME filename (timestamp) for the cropped image
            crop_filename = img_name  # same name as original
            crop_path = os.path.join(save_folder, crop_filename)

            cv2.imwrite(crop_path, crop)

print(f"Cropped images saved to: {save_folder}")
```

```python
import pandas as pd
import os
from datetime import datetime, timedelta
```

```python
cropped_images_folder = "/content/drive/MyDrive/Cropped_Lake_Images"  # change if your cropped images folder has a different name
excel_path = "/content/drive/MyDrive/April_Lake_Level_new.xlsx"        # path to your uploaded Excel file
output_csv_path = "/content/drive/MyDrive/image_water_levels.csv"      # output CSV path
```

```python
Read Excel and combine date+hour into a datetime
df_excel = pd.read_excel(excel_path)

# Assuming your Excel has columns like "date" and "hour"
df_excel['datetime'] = pd.to_datetime(df_excel['Date'].astype(str) + ' ' + df_excel['Time'].astype(str), format='%Y-%m-%d %H:%M')
df_excel = df_excel[['datetime', 'Water_Level']]

Get cropped image paths and extract timestamps
image_data = []
for fname in os.listdir(cropped_images_folder):
    if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        # Extract timestamp from filename
        # Filename format: "YYYY-MM-DD HH-MM-SS.jpg"
        timestamp_str = fname.rsplit('.', 1)[0]  # remove extension
        img_dt = datetime.strptime(timestamp_str, '%Y-%m-%d %H-%M-%S')

        image_data.append({
            'image_path': os.path.join(cropped_images_folder, fname),
            'timestamp': img_dt
        })

df_images = pd.DataFrame(image_data)
```

```python
#Match each image to nearest hour in Excel
matched_data = []
for _, row in df_images.iterrows():
    img_time = row['timestamp']
    # Find nearest datetime from Excel
    nearest_idx = (df_excel['datetime'] - img_time).abs().idxmin()
    water_level = df_excel.loc[nearest_idx, 'Water_Level']
    matched_data.append([row['image_path'], water_level])
```

```python
# Save to CSV
df_output = pd.DataFrame(matched_data, columns=['image_path', 'water_level'])
df_output.to_csv(output_csv_path, index=False)

print(f"âœ… CSV saved at: {output_csv_path}")
print(df_output.head())
```

```python
import pandas as pd
import os
from datetime import datetime

#  1. Load and combine all Excel files
excel_files = [
    "/content/drive/MyDrive/Feb_Lake_Level_Formatted_Manual (1).xlsx",
    "/content/drive/MyDrive/March_Lake_Level_Formatted.xlsx",
    "/content/drive/MyDrive/April_Lake_Level_new.xlsx"
]

df_list = []
for file in excel_files:
    temp = pd.read_excel(file)
    # Combine Date + Time into datetime
    temp["datetime"] = pd.to_datetime(temp["Date"].astype(str) + " " + temp["Time"].astype(str))
    df_list.append(temp[["datetime", "Water_Level"]])   # <-- fixed column name

# Combine all months into one DataFrame
lake_df = pd.concat(df_list).sort_values("datetime").reset_index(drop=True)

#  2. Go through cropped images
image_folder = "/content/drive/MyDrive/Cropped_Lake_Images"   # change to your folder path
output_data = []

for filename in os.listdir(image_folder):
    if filename.lower().endswith((".png", ".jpg", ".jpeg")):
        try:
            # Extract datetime from filename e.g. "2025-03-01 15-46-13.png"
            ts_str = filename.rsplit(".", 1)[0]   # remove extension
            ts = datetime.strptime(ts_str, "%Y-%m-%d %H-%M-%S")

            # Find nearest hour in lake_df
            nearest_idx = (lake_df["datetime"] - ts).abs().idxmin()
            nearest_row = lake_df.loc[nearest_idx]

            output_data.append({
                "image_path": os.path.join(image_folder, filename),
                "timestamp": ts,
                "nearest_hour": nearest_row["datetime"],
                "water_level": nearest_row["Water_Level"]   # <-- fixed here too
            })
        except Exception as e:
            print(f"Skipping {filename}, error: {e}")

# 3. Save to CSV
output_df = pd.DataFrame(output_data)
output_df.to_csv("/content/drive/MyDrive/image_water_mapping.csv", index=False)

print(" Done! CSV saved in Google Drive as image_water_mapping.csv")
print(output_df.head())
```
